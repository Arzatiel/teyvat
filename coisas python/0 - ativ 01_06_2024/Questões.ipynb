```python
'''
Questão 1) Crie um programa com while, que irá imprimir valores entre 1 a 20
'''

v = 1
while v >= 1 and v <= 20:
    print(v)
    v = (v+1)
```

```python
'''
Questão 2) Dado as senhas abaixo, crie um programa em que se o usuário acertar
uma das senhas,
aparece que acertou a senha.
Caso contrário, pedirá para o usuário digitá-la novamente.
senhas = ['certo', '123', '456']
'''
while True:
    senhas = ['certo', '123', '456']
    s = input('Digite a senha: ')
    if s in senhas:
        print('Senha Correta.')
        break
    else:
        print('Senha Inválida. \nTente Novamente.')
```

```python
'''
Questão 4) Assim como o primeiro exercício, o usuário irá digitar uma
senha, se for uma senha certa, deverá aparecer a mensagem de senha
correta. Se for errada, o usuário digitar novamente. No entanto, se for
errado 3 vezes, irá aparecer a mensagem de senha bloqueada.
'''


senhas = ['certo', '123', '456']
tent = 0
while tent < 3:
    s = input('Digite a senha: ')
    if s in senhas:
        print('Senha correta. ')
        break
    else:
        print('Senha inválida. ')
    tent += 1
if tent == 3:
    print('Tentativas excendidas. ')
```

```python
'''
Questão 5) Crie um programa que solicita do usuário a quantidade de
dígitos. Posteriormente, deverá aparecer uma senha aleatória com a
quantidade de dígitos numéricos inseridos
'''
import random
while True:
    try:
        digt = int(input('Digite a quantidade de dígitos para a senha: '))
        if digt <= 0:
            print('Digite um número maior que zero.')
        else:
            senha = ''.join([str(random.randint(0, 9)) for i in range(digt)])
            print(f'A senha é: {senha} ')
            break
    except ValueError:
        print('Por favor, digite um número válido.')
```

```python
'''
Questão 6) Agora, crie um programa que o usuário deverá informar
quantidade de letras, caracteres (como $, #, @, &...) e números e será
fornecido uma senha aleatória.
Exemplo:
senha: aA$6$123k
'''



import random
import string
while True:
    try:
        letras = int(input('Digite a quantidade de letras desejada: '))
        carac = int(input('Digite a quantidade de caracteres especiais desejada: '))
        num = int(input('Digite a quantidade de números desejada: '))
        if letras < 0 or carac < 0 or num < 0:
            print('Digite algo válido / maior que 0. ')
        else:
            senha = ''
            senha += ''.join(random.choice(string.ascii_letters) for i in range(letras))
            senha += ''.join(random.choice(string.punctuation) for i in range(carac))
            senha += ''.join(random.choice(string.digits) for i in range(num))
            senha_lista = list(senha)
            random.shuffle(senha_lista)
            senha = ''.join(senha_lista)
            print("A senha gerada é:", senha)
            break
    except ValueError:
        print("Por favor, digite números válidos.")
```
